```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def funcionx(x):
    f = x + (2 * math.sin(2 * math.pi * x)) + 4 * math.cos(2 * math.pi * x)
    return f
```

```python
tabla = np.linspace(-10*math.pi, 10*math.pi, 1000)
```

```python
proyeccion = []
for var in range (0,1000):
    proyeccion.append(funcionx(tabla[var]))
    
```

```python
ax = plt.subplot(111, projection='polar')
ax.plot(tabla, proyeccion)
ax.set_rmax(40)
ax.set_rticks([0.5, 1, 1.5, 2])  # Less radial ticks
ax.set_rlabel_position(-22.5)  # Move radial labels away from plotted line
ax.grid(True)

ax.set_title("A line plot on a polar axis", va='bottom')
plt.show()
```
