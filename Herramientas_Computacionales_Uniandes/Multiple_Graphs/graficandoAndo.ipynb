```python
import matplotlib.pyplot as plt
import numpy as np
import math
%pylab inline
```

```python
x = np.linspace(-2,2,1000)
y = (np.e**x)*np.sin(x)
prim = (x)
ter = (x + (x**2) + (x**3)/3)
cuar = (x + (x**2) + (x**3/3) - (x**5)/30)
dif1 =  y- prim
dif3 =  y- ter
dif4 =  y- cuar
```

```python
fig, axs = plt.subplots(2,3, figsize =  (18,10))

axs[0,0].plot(x,y,c='blue', label = 'e^xsin(x)')
axs[0,0].plot(x,x, c='yellow', label = 'x')
axs[0,0].set_xlabel("$x$")
axs[0,0].set_ylabel("$f(x)$")
axs[0,0].legend()
axs[0,0].set_title('Taylor $1^{er}$ grado')

axs[0,1].plot(x,y,c='blue', label = 'e^xsin(x)')
axs[0,1].plot(x,x, c='yellow', label = 'x')
axs[0,1].plot(x,ter, c='green', label = 'x + x^2 + x^3/3')
axs[0,1].set_xlabel("$x$")
axs[0,1].set_ylabel("$f(x)$")
axs[0,1].legend()
axs[0,1].set_title('Taylor $3^{er}$ grado')


axs[0,2].plot(x,y,c='blue', label = 'e^xsin(x)')
axs[0,2].plot(x,x, c='yellow', label = 'x')
axs[0,2].plot(x,ter, c='green', label = 'x + x^2 + x^3/3')
axs[0,2].plot(x,cuar, c='purple', label = 'x + x^2 + x^3/3 - x^5/30 ')
axs[0,2].set_xlabel("$x$")
axs[0,2].set_ylabel("$f(x)$")
axs[0,2].legend()
axs[0,2].set_title('Taylor $4^{to}$ grado')


axs[1,0].plot(x,dif1, c='yellow', label = 'e^xsin(x)-x')
axs[1,0].set_xlabel("$x$")
axs[1,0].set_ylabel("$f(x)$")
axs[1,0].legend()
axs[1,0].set_title('Diferencia Taylor $1^{er}$ grado')

axs[1,1].plot(x,dif3, c='green', label = 'e^xsin(x)-x')
axs[1,1].set_xlabel("$x$")
axs[1,1].set_ylabel("$f(x)$")
axs[1,1].legend()
axs[1,1].set_title('Diferencia Taylor $3^{er}$ grado')

axs[1,2].plot(x,dif4, c='purple', label = 'e^xsin(x)-x')
axs[1,2].set_xlabel("$x$")
axs[1,2].set_ylabel("$f(x)$")
axs[1,2].legend()
axs[1,2].set_title('Diferencia Taylor $4^{er}$ grado')

plt.savefig('grafica.png')
plt.show()

```
