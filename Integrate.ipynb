```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import scipy.integrate as integrate
from enum import Enum
```

```python
#Constantes:
lista = np.linspace(-2.5, 2.5, 20)
reales=[]
calculados = []
```

```python
def f(v):
    return np.e**(-v**2)
```

```python
x = np.linspace(-2.5,2.5,31416) 
exp = f(x)
plt.plot(x,exp, label = "Función", linewidth = 2, linestyle = "--")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Función")
plt.show()
```

```python
minx, maxx = -2.5, 2.5
miny, maxy = 0, np.amax(exp)
```

```python
rx = np.random.rand(20000)*(maxx-minx)+minx
ry = np.random.rand(20000)*(maxy-miny)+miny
```

```python
plt.scatter (rx,ry,alpha = 0.2, color = "blue")
plt.xlim(-2.5,2.5)
plt.ylim(0, maxy)
plt.xlabel("rx")
plt.ylabel("ry")
plt.title("Random Distribution")
plt.show()
```

```python
var = np.where (f(rx)-ry > 0.0)
plt.scatter(rx[var], ry[var], alpha = 0.5)
plt.ylim(miny, maxy+0.05)
plt.xlim(minx, maxx)
plt.xlabel("rx")
plt.ylabel("ry")
plt.show()
```

```python

def antiEva(x):
    if x <0:
        l = ( f(x)-miny )*(x-minx)
    else:
        l = ( maxy-miny )*(x-minx)
    return l*(np.size(var)/(1.0*np.size(ry)))

def lAntiEvas ():
    if np.size(calculados)>0:
        del calculados[:]
    for v in lista:
        calculados.append(antiEva(v))

    return calculados

def integralesReales ():
    if np.size(reales)>0:
        del reales[:]
    for var in lista:
        a, basura = integrate.quad(f , -2.5, var)
        reales.append(a)
    return reales
```

```python
plt.plot(lista, lAntiEvas(), label = "Calculados", linewidth = 2, linestyle = "--")
plt.plot(lista, integralesReales(), label = "Reales", linewidth = 2, linestyle = "--")
plt.legend()
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Intrgrals")
plt.show()
```

```python
def error (lt, le):
    lf = [0]
    if np.size(lt)> np.size(le):
        r = np.size(le)
    else:
        r = np.size(lt)
    for var in range (1,r):
        lf.append( ( (lt[var]-le[var])/lt[var] )*100 )
    return lf
def errorMenorQue (lt, le, x):
    l = error (lt, le)
    verificador = True
    for var in l:
        if var>x:
            verificador = False
    return verificador

print(errorMenorQue (integralesReales(),lAntiEvas() , 1))
```
